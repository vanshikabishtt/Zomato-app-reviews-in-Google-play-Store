```python
import nltk
from nltk.corpus import stopwords
sw_list = stopwords.words('english')
```

```python
df['review_text'] = df['review_text'].apply(lambda x:[item for item in x.split() if item not in sw_list]).apply(lambda x:" ".join(x))
```

```python
x = df.iloc[: , 0:1]
y = df['rating']
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
y=  encoder.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=42)
```

```python
x_train.shape , y_train.shape , x_test.shape , y_test.shape
```

```python
x_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer()
```

```python
x_train1 = cv.fit_transform(x_train['review_text']).toarray()
x_test1 = cv.transform(x_test['review_text']).toarray()
```

```python
x_train1.shape , x_test1.shape
```

```python
from sklearn.naive_bayes import MultinomialNB , GaussianNB , BernoulliNB
from sklearn.metrics import classification_report , confusion_matrix , accuracy_score
```

```python
def model(model):
  model.fit(x_train1, y_train)
  y_pred = model.predict(x_test1)
  print(classification_report(y_test , y_pred))
  print(confusion_matrix(y_test , y_pred))
  print(accuracy_score(y_test , y_pred))
```

```python
model(MultinomialNB())
```

```python
model(GaussianNB())
```

```python
model(BernoulliNB())
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model(RandomForestClassifier())
```
